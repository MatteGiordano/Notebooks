## Data Cleaning: Historical Trends in Patents

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../csv/annual_historical.csv')
df
```

```python
df.columns
```

```python
transposed_df = df.T
transposed_df = transposed_df.set_axis(transposed_df.iloc[0], axis=1)
transposed_df = transposed_df[1:]
```

```python
transposed_df.reset_index(inplace=True)
```

```python
transposed_df['subcategory_code'] = transposed_df['index'].str.extract('(\d+)')
```

```python
transposed_df['category_code'] = transposed_df['index'].str.extract('(\d)')
```

```python
transposed_df
```

```python
transposed_df["category_code"] = transposed_df["category_code"].astype("Int64")
transposed_df["subcategory_code"] = transposed_df["subcategory_code"].astype("Int64")
```

```python
nber_categories = pd.read_csv("../csv/subcategories.csv")
```

```python
nber_categories = nber_categories.drop(columns="CatNameShort")
```

```python
nber_categories.info()
```

```python
merged_df = pd.merge(left=transposed_df,right=nber_categories,how='left',left_on=["subcategory_code","category_code"],right_on=["SubcategoryCode","Category"])
merged_df.drop(columns=["SubcategoryCode","Category"], inplace=True)
```
